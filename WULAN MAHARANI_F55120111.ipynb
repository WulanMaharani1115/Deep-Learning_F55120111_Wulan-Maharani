```python
from keras.applications.vgg16 import VGG16

model = VGG16()
model.summary()
```

```python
yh = model.predict(imgs)
for i in range(len(urls)):
  if(np.argmax(yh[i]) == 436  or np.argmax(yh[i]) == 565 or np.argmax(yh[i]) == 569 or np.argmax(yh[i]) == 705 or np.argmax(yh[i]) == 751 or np.argmax(yh[i]) == 817 or np.argmax(yh[i]) == 829 or np.argmax(yh[i]) == 864):
    print("Mobil")
  elif(np.argmax(yh[i]) == 670 or np.argmax(yh[i]) == 671):
    print("Motor")
  else:
    print("Lainnya")
```
